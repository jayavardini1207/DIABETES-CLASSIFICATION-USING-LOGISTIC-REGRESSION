 Clinical Risk Assessment: Diabetes Classification using Logistic Regression

This project implements a supervised machine learning architecture utilizing **Logistic Regression** to analyze multi-dimensional clinical parameters. The primary objective is to evaluate patient data to classify and predict individual diabetes risk profiles (Binary Classification: 0 or 1).

 Dataset Architecture & Feature Descriptions

The predictive model is trained and evaluated on a clinical repository containing **768 total patient records**. The system analyzes 8 distinct independent physiological parameters to determine the final diagnostic classification.

 Independent Feature Metrics (X):

1.Pregnancies: Count of pregnancies, tracking metabolic variations correlated with reproductive history.
2.Glucose: 2-hour plasma glucose concentration resulting from an Oral Glucose Tolerance Test (OGTT). This serves as the primary clinical indicator for systemic insulin resistance.
3.BloodPressure: Diastolic blood pressure reading (measured in mm Hg), capturing vascular health when the heart rests between beats:
  a)Normal Range: Below 80 mm Hg
  b)Hypertensive Thresholds: 80 mm Hg and above (strongly linked to accelerated diabetic complications).
4.Skin Thickness: Triceps skinfold thickness measured in millimeters (mm), used as a structural metric to evaluate subcutaneous body fat distribution.
5.Insulin: 2-Hour serum insulin levels (mu U/ml). Elevated baselines indicate physiological compensation or active therapeutic management.
6.BMI (Body Mass Index): An internationally recognized structural index mapping body mass relative to height, serving as a primary marker for obesity:
7.DiabetesPedigreeFunction: A specialized scoring system that quantifies the hereditary and genetic risk of diabetes based on family lineage history.
8.Age: Chronological age of the patient in years, capturing long-term metabolic degradation over time.

 Target Classification Label (y):
a).`1` = Diagnosed Diabetic Risk Profile
b).`0` = Healthy Control Profile

  Model Evaluation & Statistical Scorecard

The data split was executed using a controlled execution seed (`random_state=2529`), preserving a 70/30 split ratio (537 training records and 231 evaluation records). This controlled shuffle ensures reproducible partitions for validating the classifier's performance.

 Overall Predictive Accuracy
Our model's performance metrics evaluate the percentage of correctly classified instances across the entire test subset. This baseline accuracy score tracks the global effectiveness of our calculated regression coefficients.
 an accuracy of 77% was predicted.

 The Diagnostic Confusion Matrix Breakdown
To track exactly how the model categorized its predictions against the actual medical ground truths established by clinical physicians, we map the final 231 test outcomes:

a).True Negatives (TN): The count of healthy individuals correctly identified by the model as non-diabetic.
b).True Positives (TP): The count of diabetic patients correctly flagged by the model's risk threshold.
c).False Positives (FP): Cases where the model raised a false alarm, incorrectly predicting diabetes for a healthy profile.
d).False Negatives (FN): Critical miss cases where a high-risk patient was incorrectly classified as healthy.

  Core Analytical Insight
The **principal diagonal** captures our model's absolute success points where the AI prediction matches clinical reality. By analyzing where the off-diagonal errors occur (False Positives vs. False Negatives), we can evaluate the clinical safety of the model. For safety-critical diagnostic applications, future iterations would focus on tuning the internal decision threshold to optimize the recall metric, minimizing any instances where a high-risk patient is falsely classified as healthy.

```python
diabetic=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
# defining or assigning the y(dependent variable) and X(independent variable)
y=diabetic['diabetes']
```

```python
y
```

```python
# X drops the column 'diabetes' and axis=1 variable is for instructing the system to perform operations on a column
X=diabetic.drop(['diabetes'],axis=1)
```

```python
X
```

```python
# implement the train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
```

```python
# checking the shape(dimensions) of the train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# using logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=600)
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
# cheking the model's accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
